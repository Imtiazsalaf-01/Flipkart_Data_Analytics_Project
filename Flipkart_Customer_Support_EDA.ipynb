# **Project Summary -**

This project focuses on analyzing Flipkart’s customer support data to gain meaningful insights into customer issues, agent performance, and overall service quality. Customer support plays a crucial role in maintaining customer satisfaction and brand loyalty in the e-commerce industry. By analyzing customer interaction data, organizations can identify recurring issues, operational inefficiencies, and opportunities for improvement. The primary objective of this project is to understand the factors that influence customer satisfaction and to provide actionable insights that can help Flipkart enhance its support operations.

The dataset used in this project contains detailed records of customer support interactions, including interaction channels, issue categories, sub-categories, order details, agent hierarchy, handling time, and customer satisfaction scores (CSAT). Each record represents a single customer interaction, making the dataset suitable for analyzing patterns and trends in customer behavior and support performance. The CSAT score serves as a key indicator of service quality and customer experience.

The exploratory data analysis began with understanding the structure of the dataset, checking data types, identifying missing values, and verifying the presence of duplicate records. Several columns were found to contain missing values, particularly in customer remarks, item price, and handling time. To ensure data quality, numerical missing values were handled using median imputation, which minimizes the effect of outliers. Columns with excessive missing values were removed to improve consistency and reliability.

Various visualizations were created to uncover insights from the data. Channel-wise analysis showed that inbound interactions are more frequent than outbound interactions, indicating that customers mostly reach out for support when they face issues. Category-wise analysis revealed that order-related and product-related issues account for the majority of customer queries, highlighting key operational areas that require attention. Analysis of CSAT scores showed that most customers provided high satisfaction ratings, reflecting a generally positive customer experience.

Agent performance analysis provided valuable insights into how experience and management influence customer satisfaction. Agents with higher tenure tended to receive better CSAT scores, suggesting that experience plays an important role in effective issue resolution. Shift-wise analysis indicated that the morning shift handles a larger volume of interactions, emphasizing the need for proper staffing during peak hours. Additionally, supervisor and manager-wise analysis showed variations in team performance, pointing to the impact of leadership and management practices.

The relationship between handling time and customer satisfaction was also examined. The analysis suggested that longer handling times can slightly reduce CSAT scores, emphasizing the importance of efficient issue resolution. Correlation analysis further revealed that customer satisfaction is influenced by multiple factors rather than a single numerical variable.

Overall, this EDA project provides a comprehensive understanding of Flipkart’s customer support operations. The insights derived from this analysis can help improve agent training programs, optimize staffing strategies, reduce recurring customer issues, and enhance overall customer satisfaction. These findings lay a strong foundation for further predictive modeling and data-driven decision-making.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python

# Data Wrangling: Making dataset analysis-ready


# 1. Remove extra spaces from column names (important)
df.columns = df.columns.str.strip()

# 2. Handling missing values for numerical columns
if 'Item_price' in df.columns:
    df['Item_price'] = df['Item_price'].fillna(df['Item_price'].median())

if 'connected_handling_time' in df.columns:
    df['connected_handling_time'] = df['connected_handling_time'].fillna(
        df['connected_handling_time'].median()
    )

# 3. Dropping columns with too many missing values (safe method)
df = df.drop(columns=['Customer Remarks'], errors='ignore')

# 4. Final check
df.isnull().sum()
````

### What all manipulations have you done and insights you found?

Missing numerical values were handled using median imputation to avoid skewness. Columns with excessive missing data were removed. These steps improved data quality and made the dataset analysis-ready.
